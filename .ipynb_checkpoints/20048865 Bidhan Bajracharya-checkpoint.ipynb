# Movie Recommendation System using Collaborative Filtering user-based model

### Importing packages

```python
# Imports for data processing
import pandas as pd
import numpy as np
import scipy.stats
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import confusion_matrix

# Import for visualization
import seaborn as sns
```

### Importing datasets

```python
# Ratings dataset
ratings = pd.read_csv('dataset/ratings.csv')
ratings.head(6)
```

```python
# checking for null values in ratings
ratings.info()
```

```python
# Movies dataset
movies = pd.read_csv('dataset/movies.csv')
movies.head(6)
```

### Merging datasets

```python
# merging ratings and movies dataset
mr = pd.merge(ratings, movies, on="movieId", how="inner")

# user ratings and movie detail in same dataset
mr.head()
```

### Exploring the dataset

```python
# Aggregate by movie
agg_ratings = mr.groupby('title').agg(mean_rating = ('rating', 'mean'), number_of_ratings = ('rating', 'count')).reset_index()
```

```python
# Visulization
sns.jointplot(x='mean_rating', y='number_of_ratings', data=agg_ratings)
```

This visualization shows that most of the movies have number of ratings less than 150 and majority of the movies has ratings between 0 to 100. <br>
We will be considering movies which has more than 50 ratings.

```python
# Number of unique ratings
sns.countplot(x='rating', data=mr, order=mr['rating'].value_counts().index)
```

As we can observe, maximum number of movies has the rating of 4 with count more than 25000, and with the count of less than 5000 the least rated movie score is 0.5. 

```python
# movies with over 50 ratings
agg_ratings_GT50 = agg_ratings[agg_ratings['number_of_ratings'] > 50]

# merging the data
df_GT50 = pd.merge(mr, agg_ratings_GT50[['title']], on='title', how='inner')
df_GT50.info()
```

```python
# total number of users
print('The dataset contains', df_GT50['userId'].nunique(), 'unique users.')

# total number of movies
print('The dataset contains', df_GT50['movieId'].nunique(), 'unique movies.')

# total unique number of ratings
print('The dataset contains', df_GT50['rating'].nunique(), 'unique ratings.')
```

### Creating user-item matrix

The matrix contains users as rows and the movie titles as columns. The values of the matrix are ratings for the movies given by respective users.

```python
# Creating user-movie matrix
user_item_matrix = df_GT50.pivot_table(index="userId", columns="title", values="rating")

# checking the data
user_item_matrix.head()
```

### Normalizing user-movie matrix

```python
# Normalizing the matrix
normalized_matrix = user_item_matrix.subtract(user_item_matrix.mean(axis=1), axis='rows')
normalized_matrix.head()
```

### Calculating similarity between users

The similarity between each user is calculated using Pearson correlation coefficient. The created table shows, the similarity score between each users. <br>
We can also observe that the similarity score of a user with itself is 1.

```python
# Calculating similarity using pearson correlation
similarity_users = normalized_matrix.T.corr()
similarity_users.head()
```

### Selecting a user to recommend movies

A user is selected at random and is removed from the dataset. <br>
The recommendation will be made for the randomly selected user.

```python
# Finding similar users

# picking active user
active_userid = 122

# removing the activer user from the list
similarity_users.drop(index=active_userid, inplace = True)

# checking
similarity_users.loc[120:125]
```

### Getting similar users

Top 10 similar users with respect to our previously selected active users are calculated here. <br>
To be considered similar with our active user, the similarity rating should be greater than certain threshold.

```python
# Similar users needed
n = 10

# Similarity threshold
similarity_threshold = 0.3

# Getting similar users
similar_users = similarity_users[similarity_users[active_userid] > similarity_threshold][active_userid].sort_values(ascending=False)[:n]

print(f'Similar top 10 users for our active user {active_userid} are:\n', similar_users)
```

### Narrowing down the item pool

The movies that our active user has seen does not need to be recommended, so it will be removed from the dataset. <br>
Similarly, those movies that none of our similar users has watched will also be removed. So, the recommendation will be made from the movies that atleast one of the similar users have watched.

```python
# Movies that are already seen by our active user
active_userid_seen = normalized_matrix[normalized_matrix.index == active_userid].dropna(axis=1, how='all')
active_userid_seen.head()
```

```python
# Removing movies that none of the similar users have watched
similar_users_watched = normalized_matrix[normalized_matrix.index.isin(similar_users.index)].dropna(axis=1, how='all')
```

```python
# Removing the active users's already watched movies from the movies that similar users have watched
similar_users_watched.drop(active_userid_seen.columns, axis=1, inplace=True, errors='ignore')
similar_users_watched.head()
```

### Making recommendations

Top 10 movies will be recommended to our active user, based on their scores. The score is determined using the weighted average of user similarity score and the movie rating. Additionally, to give more weight to the ratings of users who are more similar with our active user, users with higher similarity will get higher movie rating weight.

```python
# Storing movie scores
movie_scores = {}

for i in similar_users_watched.columns:
    # ratings for 'i' movie
    movie_rating = similar_users_watched[i]
    
    # total scores
    total = 0
    
    # sum of absolute similarity scores
    weight = 0
    
    for j in similar_users.index:
        # if movie has a rating
        if pd.isna(movie_rating[j]) == False:
            score = similar_users[j] * movie_rating[j]
            # Adding score to the total score for the movie
            total += score
            weight += abs(similar_users[j])
            
    # average score of the item
    movie_scores[i] = total/weight

# dictionary to dataframe of pandas
movie_scores = pd.DataFrame(movie_scores.items(), columns=['movie', 'movie_score'])

# sorting the movie and getting top 10
ranked_item_score = movie_scores.sort_values(by='movie_score', ascending=False)
ranked_item_score.head(10)
```

Hence, these are the recommended movies for our activer user with userId: 122.

### Calculating Accuracy Scores

```python
# Importing training and testing dataset
train_ds = pd.read_csv("ml-100k/u1.base", sep="\t", header=None, names=["userId", "movieId", "rating", "timestamp"])
test_ds = pd.read_csv("ml-100k/u1.test", sep="\t", header=None, names=["userId", "movieId", "rating", "timestamp"])
```

```python
# checking values
train_ds.head()
```

```python
# creating user-item matrix
ratings_matrix = pd.pivot_table(train_ds, values='rating', index='userId', columns='movieId')
```

```python
# normalizing the user-item matrix
normalized_matrix = ratings_matrix.subtract(ratings_matrix.mean(axis=1), axis=0)
```

```python
# calculating similarity scores
similarity_matrix = ratings_matrix.T.corr()
similarity_matrix.head()
```

### Calculating RMSE score 

```python
def calculate_rmse(u, i):
    # Checking if the item is in training datset
    # return average rating if not
    if i not in ratings_matrix.columns:
        return 2.5
    
    similarity_scores = similarity_matrix[u].drop(labels=u)
    normalized_ratings = normalized_matrix[i].drop(index=u)
    
    # Removing users that have not rated the movies
    similarity_scores.drop(index=normalized_ratings[normalized_ratings.isnull()].index, inplace=True)
    normalized_ratings.dropna(inplace=True)
    
    # Returning baseline value if, no other users has rated movies in common with active user
    if similarity_scores.isna().all():
        return 2.5
    
    total_score = 0
    total_weight = 0
    for j in normalized_ratings.index:        
        # if similarity score is not null
        if pd.isna(similarity_scores[j]) == False:
            total_score += normalized_ratings[j] * similarity_scores[j]
            total_weight += abs(similarity_scores[j])
            
    avg_user_rating = ratings_matrix.T.mean()[u]
    return avg_user_rating + total_score / total_weight
```

```python
# RMSE score of the system
test_ratings = np.array(test_ds["rating"])
user_movie_pair = zip(test_ds["userId"], test_ds["movieId"])

predicted_ratings = []
for userId, movieId in user_movie_pair:
    pred_rating = calculate_rmse(userId, movieId)
    predicted_ratings.append(pred_rating)
predicted_ratings = np.array(predicted_ratings)

print(np.sqrt(mean_squared_error(test_ratings, predicted_ratings)))
```

```python
# baseline rating RMSE score
baseline_rating = train_ds["rating"].mean()
baseline_ratings = np.array([baseline_rating for _ in range(test_ds.shape[0])])
print(np.sqrt(mean_squared_error(test_ratings, baseline_ratings)))
```

### Calculating MAE score

```python
# MAE score

# actual rating
actual_rating = test_ds['rating'].values

# predicted value
predicted_rounded = np.around(predicted_ratings)

mae = mean_absolute_error(actual_rating, predicted_rounded)
print(f'The MAE score is: {mae}')
```

### Confusion matrix ( Multiple Cases )

```python
# confusion matrix
y_actual = actual_rating
y_predicted = predicted_rounded
confusion_matrix(y_actual, y_predicted)
```
